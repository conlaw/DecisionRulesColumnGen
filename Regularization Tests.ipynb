```python
import pandas as pd
import numpy as np
import time
from Classifier import Classifier
```

```python
params = {
                'ruleGenerator': 'Hybrid',
                'masterSolver':'barrierCrossover',
                'numRulesToReturn': 100,
                'fairness_module': 'unfair'
            }
```

```python
#Equalized Odds
results = []
datasets = ['compas']

for dataset in datasets:
    print('**** DATASET %s ******'%dataset)
    for i in range(1):
        print('***** FOLD %d ******'%i)
        train  = pd.read_csv('data/split_data/'+dataset+'_train_%d.csv'%i).assign(train = True)
        test = pd.read_csv('data/split_data/'+dataset+'_test_%d.csv'%i).assign(train = False)
        
        full = pd.get_dummies(train.append(test))
        
        train = full.query('train').drop('train',axis=1)
        test = full.query('train == False').drop('train',axis=1)

        for C in [5, 10, 15, 20, 25, 30, 35, 40, 50, 100]:
            print('C: %f'%C)
            classif = Classifier(train.drop('Y',axis=1).to_numpy(), train['Y'].to_numpy(), 
                                 params, ruleGenerator = 'Generic',
                                master_model = 'compact_oneside')
            
            
            start_time = time.time()
            classif.fit(verbose = False, 
                        timeLimit = 300, 
                        timeLimitPricing = 45)    
            end_time = time.time() - start_time

            
            res = {}
            res['acc'] = sum(classif.predict(test.drop('Y',axis=1).to_numpy()) == test['Y'].to_numpy())/len(test['Y'])
            res['num_rules'] = len(classif.fitRuleSet)
            res['total_terms'] = np.sum(classif.fitRuleSet)
            res['complexity'] = len(classif.fitRuleSet) + np.sum(classif.fitRuleSet)
            res['fold'] = i
            res['data_set'] = dataset
            res['C'] = C
            res['method'] = 'compact_oneside'
            res['train_time'] = end_time
            results.append(res)
pd.DataFrame.from_records(results).to_csv('complex_constraint.csv', index=False)
```
