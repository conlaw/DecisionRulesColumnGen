```python
import pandas as pd
import numpy as np
import warnings
from binerizer import *
warnings.filterwarnings("ignore")
```

```python
fico = pd.read_csv('data/fico_clean.csv')
fico.head()
```

```python
monthCols = [col for col in fico.columns if 'MSince' in col ]
maxMonthElapsed = fico[monthCols].max().max()
```

```python
Y = fico['Y']
fico = fico.drop('Y', axis=1)
fico_validrows = (fico[~(fico == -9).all(axis = 1)].
                  replace([-8, -9], np.nan).
                 replace(-7, maxMonthElapsed))

fico_validrows['nullCol'] = fico_validrows.isnull().any(axis=1)
fico_validrows['MaxDelq2PublicRecLast12M'][fico_validrows['MaxDelq2PublicRecLast12M'] > 7] = 7
fico_validrows['Y'] = Y
fico_validrows.to_csv('data/fico_clean_final.csv', index = False)
```

```python
fico_clean = pd.read_csv('data/fico_clean_final.csv')
binerized = binerizeData(fico_clean)
```
