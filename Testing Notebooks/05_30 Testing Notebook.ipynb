```python
from test_helpers2 import readData, run_fairTest, runSingleTest, TestResults
import pandas as pd
import numpy as np
import gurobipy as gp
from gurobipy import GRB
import time
from DNFRuleModel import DNFRuleModel
from MasterModel import MasterModel
from rule_generator.GeneralRuleGenerator import GeneralRuleGenerator
from fairness_modules import *
```

```python
dataInfo = {
    'file': 'data/compas.csv',
    'groupCol': 'race'
}

X, Y, group = readData(dataInfo)
```

```python
group.shape
```

```python
num_iter = 0
args = {'group': group, 'epsilon': 0}
args_master = {'epsilon': 0, 'ruleComplexity': 5, 'masterSolver':'Default'}
args_pricing = {
                'ruleGenerator': 'Hybrid',
                'masterSolver':'Default',
                'numRulesToReturn': 100,
                'fairness_module': 'HammingEqOdd',
                'epsilon': 0,
                'ruleComplexity': 5,
                'group': group
            }

fairnessModule = EqualityOfOpportunity.EqualityOfOpportunity(args)
ruleMod = DNFRuleModel(X, Y)
master = MasterModel(ruleMod, fairnessModule, args_master)
ruleGen = GeneralRuleGenerator(ruleMod, fairnessModule, args_pricing)

verbose = False

while num_iter < 4:
    results = master.solve(verbose = verbose, relax = True)
    print(results)
    print(len(results['mu']))
    print(len(results['alpha']))

    master.finalMod.write('masterWithNewConst.lp')
    ruleFlag, rules = ruleGen.generateRule(results)
    #ruleGen.ruleGen.model.write('oldIPForrmulation.lp')
    
    if ruleFlag:
        master.addRule(rules)
    num_iter += 1
```
